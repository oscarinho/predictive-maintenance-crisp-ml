# Phase 1: Business Understanding

## 1.1 Problem Definition

**Domain:** Manufacturing / Industrial Equipment Maintenance

**Problem Statement:**
- Manufacturing equipment failures lead to costly unplanned downtime
- Traditional maintenance is either reactive (fix after failure) or time-based (fixed schedules)
- Need: Predict failures BEFORE they occur to enable **proactive maintenance**

## 1.2 Business Objectives

| Objective | Description |
|-----------|-------------|
| **Primary** | Predict machine failures before they occur |
| **Secondary** | Identify failure types for targeted maintenance |
| **Tertiary** | Understand key factors contributing to failures |

## 1.3 Success Criteria

| Metric | Target | Rationale |
|--------|--------|----------|
| **Recall** | > 80% | Minimize missed failures (costly) |
| **Precision** | > 70% | Avoid excessive false alarms |
| **F1 Score** | > 75% | Balance precision and recall |
| **ROC-AUC** | > 90% | Strong discrimination ability |

## 1.4 Business Constraints

- Model must be interpretable for maintenance teams
- Predictions needed in real-time for shop floor use
- False negatives (missed failures) are more costly than false positives

# Phase 3: Data Preparation

## 3.1 Data Cleaning

```python
# Create a working copy
df_processed = df.copy()

# Drop identifier columns (not useful for prediction)
columns_to_drop = ['UDI', 'Product ID']
df_processed = df_processed.drop(columns=columns_to_drop)

print(f"Dropped columns: {columns_to_drop}")
print(f"Remaining columns: {list(df_processed.columns)}")
```

## 3.2 Feature Engineering

```python
# Feature engineering based on mechanical and thermal domain knowledge
# Goal: Transform raw sensor readings into stress, heat, and degradation indicators
# These derived features better represent real-world failure mechanisms

# 1. Temperature Difference
# Represents heat dissipation efficiency.
# A high difference may indicate friction, overload, or cooling inefficiency.
df_processed['Temp_diff'] = (
    df_processed['Process temperature [K]'] -
    df_processed['Air temperature [K]']
)

# 2. Mechanical Power Proxy
# Torque Ã— Rotational Speed approximates mechanical power demand.
# Higher values indicate heavier operational load on the machine.
df_processed['Power'] = (
    df_processed['Torque [Nm]'] *
    df_processed['Rotational speed [rpm]']
)

# 3. Mechanical Strain Indicator
# Combines instantaneous stress (torque) with accumulated degradation (tool wear).
# This captures long-term fatigue effects better than either feature alone.
df_processed['Strain'] = (
    df_processed['Torque [Nm]'] *
    df_processed['Tool wear [min]']
)

# 4. Normalized Tool Wear
# Scales tool wear to a 0â€“1 range.
# Useful for:
# - Model stability
# - Comparability across machines and time periods
df_processed['Tool_wear_ratio'] = (
    df_processed['Tool wear [min]'] /
    df_processed['Tool wear [min]'].max()
)

# 5. Temperature Ratio
# Relative thermal load indicator.
# Highlights abnormal thermal behavior independent of ambient conditions.
df_processed['Temp_ratio'] = (
    df_processed['Process temperature [K]'] /
    df_processed['Air temperature [K]']
)

# Log summary for traceability and notebook readability
print("âœ… New features created using domain knowledge:")
print("   â€¢ Temp_diff â†’ Heat dissipation efficiency")
print("   â€¢ Power â†’ Mechanical workload proxy (Torque Ã— RPM)")
print("   â€¢ Strain â†’ Combined stress & fatigue indicator")
print("   â€¢ Tool_wear_ratio â†’ Normalized degradation level")
print("   â€¢ Temp_ratio â†’ Relative thermal stress")
```

Why feature engineering matters here:
 * Raw sensor values describe measurements.
 * Engineered features describe failure mechanisms.

This step bridges maintenance engineering knowledge and machine learning, which is critical for real-world adoption.

```python
# Visualize new features
new_features = ['Temp_diff', 'Power', 'Strain', 'Tool_wear_ratio', 'Temp_ratio']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, col in enumerate(new_features):
    sns.boxplot(data=df_processed, x='Target', y=col, ax=axes[i],
                palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(f'{col} by Target', fontsize=11, fontweight='bold')
    axes[i].set_xticklabels(['No Failure', 'Failure'])

axes[5].axis('off')  # Hide empty subplot
plt.tight_layout()
plt.show()
```

Compared to raw sensor readings, the engineered features capture meaningful shifts in mechanical load, fatigue, and degradation, demonstrating why domain-driven feature engineering is essential for predictive maintenance models.

# Phase 4: Modeling

## 4.1 Baseline Model Comparison

```python
from sklearn.pipeline import Pipeline

models = {
    'Logistic Regression': Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', LogisticRegression(
            random_state=42,
            max_iter=1000,
            class_weight='balanced'
        ))
    ]),
    
    'Decision Tree': Pipeline(steps=[
        ('model', DecisionTreeClassifier(
            random_state=42,
            class_weight='balanced'
        ))
    ]),
    
    'Random Forest': Pipeline(steps=[
        ('model', RandomForestClassifier(
            random_state=42,
            n_estimators=100,
            class_weight='balanced'
        ))
    ]),
    
    'Gradient Boosting': Pipeline(steps=[
        ('model', GradientBoostingClassifier(
            random_state=42,
            n_estimators=100
        ))
    ]),
    
    'XGBoost': Pipeline(steps=[
        ('model', XGBClassifier(
            random_state=42,
            n_estimators=100,
            scale_pos_weight=28,
            eval_metric='logloss'
        ))
    ]),
    
    'K-Nearest Neighbors': Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', KNeighborsClassifier(n_neighbors=5))
    ]),
        'CatBoost': Pipeline(steps=[
        ('model', CatBoostClassifier(
            iterations=300,
            learning_rate=0.05,
            depth=6,
            loss_function='Logloss',
            class_weights=[1, 28],
            verbose=False,
            random_state=42
        ))
    ]),
    
    'LightGBM': Pipeline(steps=[
        ('model', LGBMClassifier(
            n_estimators=300,
            learning_rate=0.05,
            class_weight='balanced',
            random_state=42
        ))
    ]),
    'SVM': Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', SVC(
            random_state=42,
            probability=True,
            class_weight='balanced'
        ))
    ])
}

print(f"Comparing {len(models)} models using pipeline-safe preprocessing...\n")
```

```python
# Cross-validation and evaluation (fix XGBoost feature-name issue)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

# Use numpy arrays to avoid XGBoost feature-name restrictions ([, ], <)
X_train_np = X_train.values
X_test_np = X_test.values

for name, model in models.items():
    print(f"Training {name}...", end=" ")
    
    # Cross-validation scores
    cv_accuracy = cross_val_score(model, X_train_np, y_train, cv=cv, scoring='accuracy')
    cv_precision = cross_val_score(model, X_train_np, y_train, cv=cv, scoring='precision')
    cv_recall = cross_val_score(model, X_train_np, y_train, cv=cv, scoring='recall')
    cv_f1 = cross_val_score(model, X_train_np, y_train, cv=cv, scoring='f1')
    cv_roc_auc = cross_val_score(model, X_train_np, y_train, cv=cv, scoring='roc_auc')
    
    # Train on full training set
    model.fit(X_train_np, y_train)
    
    # Test set predictions
    y_pred = model.predict(X_test_np)
    y_pred_proba = (
        model.predict_proba(X_test_np)[:, 1]
        if hasattr(model, "predict_proba")
        else None
    )
    
    # Store results
    results[name] = {
        'model': model,
        'cv_accuracy': (cv_accuracy.mean(), cv_accuracy.std()),
        'cv_precision': (cv_precision.mean(), cv_precision.std()),
        'cv_recall': (cv_recall.mean(), cv_recall.std()),
        'cv_f1': (cv_f1.mean(), cv_f1.std()),
        'cv_roc_auc': (cv_roc_auc.mean(), cv_roc_auc.std()),
        'test_accuracy': accuracy_score(y_test, y_pred),
        'test_precision': precision_score(y_test, y_pred, zero_division=0),
        'test_recall': recall_score(y_test, y_pred),
        'test_f1': f1_score(y_test, y_pred),
        'test_roc_auc': roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba
    }
    
    print(f"âœ“ (CV F1: {cv_f1.mean():.4f})")

print("\nâœ… All models trained using pipeline-safe evaluation!")
```

```python
# Create comparison table (safe formatting after pipeline + XGBoost fix)
comparison_data = []

for name, res in results.items():
    comparison_data.append({
        'Model': name,
        'CV Accuracy': f"{res['cv_accuracy'][0]:.4f} Â± {res['cv_accuracy'][1]:.4f}",
        'CV Precision': f"{res['cv_precision'][0]:.4f} Â± {res['cv_precision'][1]:.4f}",
        'CV Recall': f"{res['cv_recall'][0]:.4f} Â± {res['cv_recall'][1]:.4f}",
        'CV F1': f"{res['cv_f1'][0]:.4f} Â± {res['cv_f1'][1]:.4f}",
        'CV ROC-AUC': f"{res['cv_roc_auc'][0]:.4f} Â± {res['cv_roc_auc'][1]:.4f}",
        'Test F1': f"{res['test_f1']:.4f}",
        'Test ROC-AUC': f"{res['test_roc_auc']:.4f}" if res['test_roc_auc'] is not None else "N/A"
    })

comparison_df = pd.DataFrame(comparison_data)
comparison_df = comparison_df.sort_values('Test F1', ascending=False)

print("\n" + "="*100)
print("MODEL COMPARISON (Cross-VALIDATION RESULTS)")
print("="*100)
display(comparison_df.reset_index(drop=True))
```

```python
# Visualize model comparison (robust + sorted)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sort models by Test F1 (descending)
sorted_models = sorted(
    results.keys(),
    key=lambda m: results[m]['test_f1'],
    reverse=True
)

# F1 Score comparison
model_names = sorted_models
f1_scores = [results[m]['test_f1'] for m in model_names]
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(model_names)))

bars = axes[0].barh(model_names, f1_scores, color=colors, edgecolor='black')
axes[0].set_xlabel('F1 Score')
axes[0].set_title('Model Comparison â€“ Test F1 Score', fontweight='bold')
axes[0].set_xlim(0, 1)

for bar, score in zip(bars, f1_scores):
    axes[0].text(
        score + 0.01,
        bar.get_y() + bar.get_height() / 2,
        f'{score:.4f}',
        va='center'
    )

# ROC Curves
for name in model_names:
    y_proba = results[name]['y_pred_proba']
    auc = results[name]['test_roc_auc']
    
    if y_proba is not None and auc is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        axes[1].plot(fpr, tpr, label=f'{name} (AUC={auc:.3f})')

axes[1].plot([0, 1], [0, 1], 'k--', label='Random')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('ROC Curves Comparison', fontweight='bold')
axes[1].legend(loc='lower right', fontsize=8)

plt.tight_layout()
plt.show()
```

## 4.2 Hyperparameter Tuning

Based on initial results, we'll tune the top 2 performing models.

```python
# Identify best models to tune (top 3 by Test F1 â€“ safe & explicit)
sorted_models = sorted(
    results.items(),
    key=lambda item: item[1]['test_f1'],
    reverse=True
)

top_models = [name for name, _ in sorted_models[:3]]

print("Top models to tune (by Test F1):")
for i, name in enumerate(top_models, start=1):
    print(f"{i}. {name}")
```

```python
# Random Forest Hyperparameter Tuning (pipeline-safe)

print("Tuning Random Forest...")

rf_param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [10, 20, 30, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__class_weight': ['balanced']
}

rf_pipeline = Pipeline(steps=[
    ('model', RandomForestClassifier(random_state=42))
])

rf_grid = GridSearchCV(
    rf_pipeline,
    rf_param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

rf_grid.fit(X_train, y_train)

print(f"\nBest parameters: {rf_grid.best_params_}")
print(f"Best CV F1 score: {rf_grid.best_score_:.4f}")
```

```python
# Gradient Boosting Hyperparameter Tuning (pipeline-safe)

print("Tuning Gradient Boosting...")

gb_param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__learning_rate': [0.01, 0.1, 0.2],
    'model__max_depth': [3, 5, 7],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}

gb_pipeline = Pipeline(steps=[
    ('model', GradientBoostingClassifier(random_state=42))
])

gb_grid = GridSearchCV(
    gb_pipeline,
    gb_param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

gb_grid.fit(X_train, y_train)

print(f"\nBest parameters: {gb_grid.best_params_}")
print(f"Best CV F1 score: {gb_grid.best_score_:.4f}")
```

```python
# LightGBM Hyperparameter Tuning (FAST)

from lightgbm import LGBMClassifier
from sklearn.model_selection import RandomizedSearchCV

print("Tuning LightGBM (fast)...")

X_train_np = X_train.to_numpy()

lgbm_param_dist = {
    'model__n_estimators': [100, 200, 300],
    'model__learning_rate': [0.05, 0.1, 0.2],
    'model__max_depth': [-1, 5, 10],
    'model__num_leaves': [31, 63],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0],
    'model__min_child_samples': [20, 50],
    'model__class_weight': ['balanced']
}

lgbm_pipeline = Pipeline(steps=[
    ('model', LGBMClassifier(
        random_state=42,
        objective='binary',
        verbosity=-1,
        force_col_wise=True
    ))
])

lgbm_search = RandomizedSearchCV(
    lgbm_pipeline,
    lgbm_param_dist,
    n_iter=20,         # << controls speed (try 10â€“30)
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=0,
    random_state=42
)

lgbm_search.fit(X_train_np, y_train)

print(f"\nBest parameters: {lgbm_search.best_params_}")
print(f"Best CV F1 score: {lgbm_search.best_score_:.4f}")
```

```python
# Compare tuned models (pipeline-safe, fixed for LightGBM RandomizedSearch)

tuned_models = {
    'Random Forest (Tuned)': rf_grid.best_estimator_,
    'Gradient Boosting (Tuned)': gb_grid.best_estimator_,
    'LightGBM (Tuned)': lgbm_search.best_estimator_   # <-- FIX HERE
}

print("\nTuned Model Performance on Test Set:")
print("="*60)

for name, model in tuned_models.items():
    # LightGBM was trained with numpy arrays
    if name.startswith("LightGBM"):
        X_eval = X_test.to_numpy()
    else:
        X_eval = X_test

    y_pred = model.predict(X_eval)
    y_pred_proba = model.predict_proba(X_eval)[:, 1]
    
    print(f"\n{name}:")
    print(f"   Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"   Precision: {precision_score(y_test, y_pred, zero_division=0):.4f}")
    print(f"   Recall:    {recall_score(y_test, y_pred):.4f}")
    print(f"   F1 Score:  {f1_score(y_test, y_pred):.4f}")
    print(f"   ROC-AUC:   {roc_auc_score(y_test, y_pred_proba):.4f}")
```

## 4.3 Final Model Selection

```python
from sklearn.metrics import f1_score

# Calcula F1 de cada modelo (pipeline-safe; sin X_test_scaled)
scores = {
    "Random Forest (Tuned)": f1_score(y_test, rf_grid.best_estimator_.predict(X_test)),
    "Gradient Boosting (Tuned)": f1_score(y_test, gb_grid.best_estimator_.predict(X_test)),
    "LightGBM (Tuned)": f1_score(y_test, lgbm_search.best_estimator_.predict(X_test.to_numpy())),
}

# Elige el nombre del mejor segÃºn F1
final_model_name = max(scores, key=scores.get)

# Mapea nombre -> buscador para extraer modelo y params
searchers = {
    "Random Forest (Tuned)": rf_grid,
    "Gradient Boosting (Tuned)": gb_grid,
    "LightGBM (Tuned)": lgbm_search,
}

final_model = searchers[final_model_name].best_estimator_
best_params = searchers[final_model_name].best_params_

print("\nF1 scores:")
for k, v in scores.items():
    print(f"  - {k}: {v:.4f}")

print(f"\nðŸ† FINAL MODEL SELECTED: {final_model_name}")
print("\nBest Parameters:")
for param, value in best_params.items():
    print(f"   â€¢ {param}: {value}")
```

## 5.4 Feature Importance Analysis

```python
# Feature importance (pipeline-safe, model-aware)

# Extract the underlying estimator from the pipeline
if hasattr(final_model, "named_steps"):
    estimator = final_model.named_steps["model"]
else:
    estimator = final_model

# Ensure the model supports feature importance
if not hasattr(estimator, "feature_importances_"):
    raise AttributeError("Final model does not expose feature_importances_")

# Build feature-importance DataFrame
feature_importance = pd.DataFrame({
    'Feature': feature_columns,
    'Importance': estimator.feature_importances_
}).sort_values('Importance', ascending=True)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

colors = plt.cm.viridis(
    feature_importance['Importance'] / feature_importance['Importance'].max()
)
bars = ax.barh(
    feature_importance['Feature'],
    feature_importance['Importance'],
    color=colors,
    edgecolor='black'
)

ax.set_xlabel('Importance Score')
ax.set_title('Feature Importance', fontsize=14, fontweight='bold')

# Add value labels
for bar, importance in zip(bars, feature_importance['Importance']):
    ax.text(
        importance + 0.005,
        bar.get_y() + bar.get_height() / 2,
        f'{importance:.4f}',
        va='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()

# Print ranking
print("\nFeature Importance Ranking:")
print("=" * 50)
for i, (_, row) in enumerate(feature_importance.iloc[::-1].iterrows(), 1):
    print(f"{i:2d}. {row['Feature']:30s} {row['Importance']:.4f}")
```

## 5.5 Threshold Analysis

```python
# Analyze different probability thresholds
thresholds = np.arange(0.1, 0.9, 0.1)
threshold_results = []

for thresh in thresholds:
    y_pred_thresh = (y_pred_proba_final >= thresh).astype(int)
    threshold_results.append({
        'Threshold': thresh,
        'Precision': precision_score(y_test, y_pred_thresh, zero_division=0),
        'Recall': recall_score(y_test, y_pred_thresh),
        'F1': f1_score(y_test, y_pred_thresh)
    })

thresh_df = pd.DataFrame(threshold_results)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(thresh_df['Threshold'], thresh_df['Precision'], 'b-o', label='Precision')
ax.plot(thresh_df['Threshold'], thresh_df['Recall'], 'g-s', label='Recall')
ax.plot(thresh_df['Threshold'], thresh_df['F1'], 'r-^', label='F1 Score')

ax.axvline(x=0.5, color='gray', linestyle='--', alpha=0.7, label='Default (0.5)')
ax.set_xlabel('Probability Threshold')
ax.set_ylabel('Score')
ax.set_title('Metrics vs Classification Threshold', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nThreshold Analysis:")
display(thresh_df.round(4))
```

This analysis shows that adjusting the classification threshold enables a deliberate trade-off between early failure detection and false alarms, allowing the model to be aligned with operational risk tolerance rather than relying on a generic 0.5 cutoff.

```python
# Business cost simulation
print("\n" + "="*60)
print("BUSINESS IMPACT ANALYSIS")
print("="*60)

# Assumed costs (can be adjusted based on actual business data)
cost_unplanned_failure = 10000  # Cost of unplanned failure (downtime + repairs)
cost_preventive_maintenance = 500  # Cost of preventive maintenance
cost_false_alarm = 200  # Cost of unnecessary maintenance check

tn, fp, fn, tp = cm.ravel()

# With model
cost_with_model = (
    fn * cost_unplanned_failure +  # Missed failures still cause downtime
    tp * cost_preventive_maintenance +  # Prevented failures with maintenance
    fp * cost_false_alarm  # False alarms trigger unnecessary checks
)

# Without model (all failures are unplanned)
total_failures = tp + fn
cost_without_model = total_failures * cost_unplanned_failure

savings = cost_without_model - cost_with_model
savings_pct = (savings / cost_without_model) * 100

print(f"""
Assumed Costs:
   â€¢ Unplanned failure: ${cost_unplanned_failure:,}
   â€¢ Preventive maintenance: ${cost_preventive_maintenance:,}
   â€¢ False alarm check: ${cost_false_alarm:,}

Cost Analysis (Test Set - {len(y_test)} machines):
   â€¢ WITHOUT Model: ${cost_without_model:,}
   â€¢ WITH Model: ${cost_with_model:,}
   â€¢ SAVINGS: ${savings:,} ({savings_pct:.1f}%)

Breakdown:
   â€¢ Failures prevented (TP): {tp} (saved ${tp * (cost_unplanned_failure - cost_preventive_maintenance):,})
   â€¢ Failures missed (FN): {fn} (cost ${fn * cost_unplanned_failure:,})
   â€¢ False alarms (FP): {fp} (cost ${fp * cost_false_alarm:,})
""")
```
